```python
from sklearn import datasets
from sklearn.linear_model import Perceptron   
from sklearn.model_selection import train_test_split   # tap train, tap test
from sklearn.metrics import accuracy_score  # lay do chinh xac
from sklearn.preprocessing import StandardScaler    # chuan hoa du lieu, chia ty le du lieu deu nhau de de dang train
import pandas as pd
```

```python
url="/content/pima-indians-diabetes.csv"
data = pd.read_csv(url)
data.head()
```

```python
dataFrame = data.drop([], axis=1).head()
```

```python
X = dataFrame.iloc[:,:4]
y = dataFrame.iloc[:,4]
```

```python
X.head()
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

```python
Sc = StandardScaler()
Sc.fit(X_train)
```

```python
X_train_std=Sc.transform(X_train)
```

```python
X_test_std=Sc.transform(X_test)
```

```python
model = Perceptron(max_iter=40,eta0=0.1, random_state=0)
# max_iter: so lan hoc, eta0: toc do hoc
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(y_test)
print(y_pred)
print("Do chinh xac khi chua scale: ", accuracy_score(y_test, y_pred))
```

```python
model = Perceptron(max_iter=40,eta0=0.1, random_state=0)
model.fit(X_train_std, y_train)
y_pred = model.predict(X_test_std)
print(y_test)
print(y_pred)
print('Do chinh xac khi scale: ', accuracy_score(y_test, y_pred))
```
